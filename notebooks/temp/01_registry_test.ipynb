```python
import numpy as np
import photon_tools as pt
from photon_tools.model import PhotonData, PhotonDataset
```

```python
def dummy_loader(path, **kw):
    photons = PhotonData(
        timestamps=np.array([10, 20, 30], dtype=np.int64),
        tick_ns=5.0,
    )
    return PhotonDataset(
        photons=photons,
        meta={"format": "dummy"},
        source=str(path),
    )
```
